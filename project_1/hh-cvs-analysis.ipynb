В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


````python
def extract_experience_months(value):
    if value is np.nan:
        return np.nan
    if value == 'Не указано':
        return np.nan

    experience_summary = value.split('  ')[0]

    summary_parts = experience_summary.lower().split()

    experience_months = 0
    while True:
        part = summary_parts.pop()
        if part.startswith('месяц'):
            experience_months = int(summary_parts.pop())
        else:
            if part.startswith('год') or part == 'лет':
                experience_months += 12 * int(summary_parts.pop())
            break

    return experience_months


# display(extract_experience_months('Опыт работы 16 лет 10 месяцев  Август 2010...'))
# display(extract_experience_months('Опыт работы 3 года 10 месяцев  Специалист 23 000...'))
# display(extract_experience_months('Опыт работы 1 год 10 месяцев  Специалист 23 000...'))
# display(extract_experience_months('Опыт работы 1 год  Специалист 23 000...'))
# display(extract_experience_months('Опыт работы 16 лет  Август 2010...'))
# display(extract_experience_months('Опыт работы 10 месяцев  Специалист 23 000...'))
# display(extract_experience_months('Опыт работы 1 месяц  Специалист 23 000...'))

hh['Опыт работы (месяц)'] = hh['Опыт работы'].apply(extract_experience_months)
display(hh['Опыт работы (месяц)'].shape)
display(hh['Опыт работы (месяц)'].describe())

hh.drop(columns='Опыт работы', inplace=True)
````

````python
def extract_city(value):
    return value[:value.index(',')].strip()


keep_cities = set(['Москва', 'Санкт-Петербург'])
million_cities = set(['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
                     'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'])


def normalize_city(value):
    if value in keep_cities:
        return value
    if value in million_cities:
        return 'город-миллионник'
    return 'другие'


def extract_relocation(value):
    for part in value.split(','):
        if 'перее' in part:
            part = part.replace(' ', '')
            if 'неготов' in part or 'нехочу' in part:
                return False
            if 'готов' in part or 'хочу' in part:
                return True
            return np.nan
    return np.nan


def extract_trip(value):
    for part in value.split(','):
        if 'командировка' in part:
            part = part.replace(' ', '')
            if 'неготов' in part:
                return False
            if 'готов' in part:
                return True
            return np.nan
    return False


hh['Город'] = hh['Город, переезд, командировки'].apply(extract_city)
hh['Готовность к переезду'] = hh['Город, переезд, командировки'].apply(extract_relocation)
hh['Готовность к командировкам'] = hh['Город, переезд, командировки'].apply(extract_trip)

# print('Уникальных городов:', hh['Город'].unique().shape)
# print('Уникальных городов в нижнем регистре после удаления спец. символов:',
#       hh['Город'].str.lower().replace(' ', '').replace('-', '').unique().shape)
# print()

hh['Город'] = hh['Город'].apply(normalize_city)

# hh[['Город', 'Готовность к переезду', 'Готовность к командировкам']].info()
# display(hh[['Город', 'Готовность к переезду', 'Готовность к командировкам']].describe())

print(int(round(hh['Город'].value_counts(normalize=True) * 100)['Санкт-Петербург']),
      'процентов соискателей живут в Санкт-Петербурге.')
print(round((hh['Готовность к переезду'] & hh['Готовность к командировкам']).mean() * 100),
      'процентов соискателей готовы одновременно и к переездам, и к командировкам')

hh.drop(columns='Город, переезд, командировки', inplace=True)
````

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

````python
def one_hot_encoding(df, category):
    # convert column with CSV list to column with set
    category_set = df[category] \
        .apply(lambda cell: {value.strip() for value in cell.split(',')})
    
    # get unique values across column with sets
    unique_values = category_set.agg(lambda column: set.union(*column))
    
    # convert unique values set to sorted list
    unique_values = list(unique_values)
    unique_values.sort()
    
    for value in unique_values:
        # for each unique value create True/False column
        df[value] = category_set.apply(lambda values: value in values)
        
    return unique_values

employment_types = one_hot_encoding(hh, 'Занятость')
schedule_types = one_hot_encoding(hh, 'График')
# hh[employment_types + schedule_types].info()
# display(hh[employment_types + schedule_types].describe())

print((hh['проектная работа'] & hh['волонтерство']).sum(), 'людей ищут проектную работу и волонтёрство')
print((hh['вахтовый метод'] & hh['гибкий график']).sum(), 'людей хотят работать вахтовым методом и с гибким графиком')

hh.drop(columns=['Занятость', 'График'], inplace=True)
````

Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


````python
def extract_amount(value):
    return int(value.split()[0])


def extract_currency(value):
    curency_name = value.split()[1].strip('.')
    curency_name_to_iso = {
        'руб': 'RUB',
        'бел.руб': 'BYN',
        'грн': 'UAH',
        'сум': 'UZS',
    }
    return curency_name_to_iso.get(curency_name, curency_name)


hh['date'] = pd.to_datetime(hh['Обновление резюме'], dayfirst=True).dt.date
hh['ЗП (руб)'] = hh['ЗП'].apply(extract_amount)
hh['currency'] = hh['ЗП'].apply(extract_currency)

rates = pd.read_csv(rates_data_location)

rates['date'] = pd.to_datetime(rates['date'], format='%d/%m/%y').dt.date
rates['rate'] = rates['close'] / rates['proportion']

hh = hh.merge(rates[['currency', 'date', 'rate']],
              how='left', on=['currency', 'date'])
hh.fillna({'rate': 1}, inplace=True)

hh['ЗП (руб)'] = hh['ЗП (руб)'] * hh['rate']

print('Медианная заработная плата соискателей в нашей таблице (в тыс. рублей):',
      round(hh['ЗП (руб)'].median() / 1000))

hh.drop(columns=['ЗП', 'currency', 'date', 'rate'], inplace=True)
````
